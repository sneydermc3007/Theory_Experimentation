```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
cantidadDados = int(input("Cuantos dados quiere lanzar"))

posibilidades = 6**cantidadDados

print(f"La cantidad de posibilidades al lanzar {cantidadDados} dados es de: ", posibilidades)
```

```python
rangoInferior = cantidadDados * 1
rangoSuperior = cantidadDados * 6

print(f"El rango es de numeros es entre: {rangoInferior} y {rangoSuperior}")

buscarDato = int(input("Seg√∫n el rango anterior que numero desea encontrar la probabilidad: "))
print(buscarDato)
```

```python
matriz = np.zeros([posibilidades, cantidadDados])
matriz_aux = []
print(len(matriz))
print(matriz)
```

```python
Pesos=np.zeros([cantidadDados])
for i in range(0,cantidadDados):
  Pesos[i]=int(6**i)

print(Pesos)
```

```python
for i in range(cantidadDados):
  contador =0
  Valor=Pesos[i]
  while contador < 6**cantidadDados:
      for p in range(6):
        for k in range(int(Valor)):
          matriz[contador, i] = p + 1
          contador=contador + 1

print(matriz)
```

```python
acumuladora, cont = [0, 0]

for i in range(posibilidades):
  cont = sum(matriz[i])

  if (buscarDato == cont):
    matriz_aux.append(matriz[i])
    acumuladora += 1

print(matriz_aux)

total = (acumuladora/posibilidades)
print(f"\nLa probabilidad de sacar {buscarDato} al tirar {cantidadDados} dados es de: {total}")

```

```python
resultado = []
numero = []
suma = 0

for i in range(rangoInferior, rangoSuperior+1):
  prueba = 0
  numero.append(i)

  for j in range(posibilidades):
    print(j)
    suma = sum(matriz[j])
    if(suma == cantidadDados):
      prueba += 1

  cantidadDados = cantidadDados + 1
  resultado.append(prueba)

print(numero)
print(resultado)

probabilidades = list(map(lambda x: x / posibilidades, resultado))
```

```python
plt.plot(numero, probabilidades, "*g")
plt.xlabel("Cantidad dados")
plt.ylabel("Probabilidad")
plt.grid()
```
