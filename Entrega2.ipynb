```python
# Dataset
import numpy as np
import matplotlib.pyplot as plt

X_1 = np.array([1, 6, 9, 12, 14, 18])
Y_1 = np.array([10, 16, 19, 32, 34, 38])

dataTamano_1 = len(X_1)

plt.plot(X_1, Y_1, '-b')
plt.title('Dataset 1')
```

```python
import numpy as np
import matplotlib.pyplot as plt

X_2 = np.array([1, 2, 3, 4, 5, 6, 7, 8])
Y_2 = np.array([21, 35, 57, 87, 125, 171, 225, 287])

dataTamano_2 = len(X_2)

plt.plot(X_2, Y_2, '-r')
plt.title('Dataset 2')
```

```python
import numpy as np
import matplotlib.pyplot as plt

X_3 = np.array([2, 2.5, 3, 4.5, 5, 6.5, 7.5, 8.5])
Y_3 = np.array([1, 4.5, 8, 14.5, 22.5, 24.5, 29.5, 34.5])

dataTamano_3 = len(X_3)

plt.plot(X_3, Y_3, '-m')
plt.title('Dataset 3')
```

```python
# Calculos estadísticos
## Media data1

AccuX_1 = 0
AccuY_1 = 0
for i in range(0, dataTamano_1, 1):
  AccuX_1=AccuX_1 + X_1[i]
  AccuY_1=AccuY_1 + Y_1[i]

MediaX_1 = AccuX_1 / dataTamano_1
MediaY_1 = AccuY_1 / dataTamano_1
print(MediaX_1)
print(MediaY_1)

## Media data2

AccuX_2 = 0
AccuY_2 = 0
for i in range(0, dataTamano_2, 1):
  AccuX_2=AccuX_2 + X_2[i]
  AccuY_2=AccuY_2 + Y_2[i]

MediaX_2 = AccuX_2 / dataTamano_2
MediaY_2 = AccuY_2 / dataTamano_2
print(MediaX_2)
print(MediaY_2)

## Media data3

AccuX_3 = 0
AccuY_3 = 0
for i in range(0, dataTamano_3, 1):
  AccuX_3=AccuX_3 + X_3[i]
  AccuY_3=AccuY_3 + Y_3[i]

MediaX_3 = AccuX_3 / dataTamano_3
MediaY_3 = AccuY_3 / dataTamano_3
print(MediaX_3)
print(MediaY_3)
```

```python
## Calcular la desviación 1
import math as M
Sumatoria_X_1 = 0
Sumatoria_Y_1 = 0
for i in range(0,dataTamano_1,1):
  Sumatoria_X_1 = Sumatoria_X_1 + (X_1[i] - MediaX_1)**2
  Sumatoria_Y_1 = Sumatoria_Y_1 + (Y_1[i] - MediaY_1)**2

Desviacion_X_1 = M.sqrt(Sumatoria_X_1/dataTamano_1)
Desviacion_Y_1 = M.sqrt(Sumatoria_Y_1/dataTamano_1)

print('La desviación estandar en el eje X_1 es:', Desviacion_X_1)
print('La desviación estandar en el eje Y_1 es:', Desviacion_Y_1)

print('=======================================')

## Calcular la desviación 2
Sumatoria_X_2 = 0
Sumatoria_Y_2 = 0
for i in range(0,dataTamano_2,1):
  Sumatoria_X_2 = Sumatoria_X_2 + (X_2[i] - MediaX_2)**2
  Sumatoria_Y_2 = Sumatoria_Y_2 + (Y_2[i] - MediaY_2)**2

Desviacion_X_2 = M.sqrt(Sumatoria_X_2/dataTamano_2)
Desviacion_Y_2 = M.sqrt(Sumatoria_Y_2/dataTamano_2)

print('La desviación estandar en el eje X_2 es:', Desviacion_X_2)
print('La desviación estandar en el eje Y_2 es:', Desviacion_Y_2)

print('=======================================')

## Calcular la desviación 3
Sumatoria_X_3 = 0
Sumatoria_Y_3 = 0
for i in range(0,dataTamano_3,1):
  Sumatoria_X_3 = Sumatoria_X_3 + (X_3[i] - MediaX_3)**2
  Sumatoria_Y_3 = Sumatoria_Y_3 + (Y_3[i] - MediaY_3)**2

Desviacion_X_3 = M.sqrt(Sumatoria_X_3/dataTamano_3)
Desviacion_Y_3 = M.sqrt(Sumatoria_Y_3/dataTamano_3)

print('La desviación estandar en el eje X_3 es:', Desviacion_X_3)
print('La desviación estandar en el eje Y_3 es:', Desviacion_Y_3)
```

```python
Cov_XY_1=np.cov(X_1,Y_1)
print('La covarianza en el Dataset 1 es:',Cov_XY_1[0,1])
print('============')

Cov_XY_2=np.cov(X_2,Y_2)
print('La covarianza en el Dataset 2 es:',Cov_XY_2[0,1])
print('============')

Cov_XY_3=np.cov(X_3,Y_3)
print('La covarianza en el Dataset 3 es:',Cov_XY_3[0,1])
print('============')
```

```python
## Cálculo de la pendiente 1
m_1=Cov_XY_1[0,1]/Desviacion_X_1**2
print('La pendiente es:',m_1)
print('============')

## Cálculo de la pendiente 2
m_2=Cov_XY_2[0,1]/Desviacion_X_2**2
print('La pendiente es:',m_2)
print('============')

## Cálculo de la pendiente 3
m_3=Cov_XY_3[0,1]/Desviacion_X_3**2
print('La pendiente es:',m_3)
print('============')
```

```python
## Cálculo de intercepción 1
b_1= MediaY_1 - m_1* MediaX_1
print('El intercepto es:',b_1)
print('============')

## Cálculo de intercepción 2
b_2= MediaY_2 - m_2* MediaX_2
print('El intercepto es:',b_2)
print('============')

## Cálculo de intercepción 3
b_3= MediaY_3 - m_3* MediaX_3
print('El intercepto es:',b_3)
print('============')
```

```python
## Modelo matematico 1
Y_N_1 = m_1 * X_1 + b_1
plt.plot(X_1, Y_1,'.k',X_1, Y_N_1,'-b')
```

```python
## Modelo matematico 2
Y_N_2 = m_2 * X_2 + b_2
plt.plot(X_2, Y_2,'.k',X_2, Y_N_2,'-b')
```

```python
## Modelo matematico 3
Y_N_3 = m_3 * X_3 + b_3
plt.plot(X_3, Y_3,'.k',X_3, Y_N_3,'-b')
```
