```python
# Importando DataSet

import pandas as pd
import numpy as np

# Cargando datos pima indians DataSet
dataset = pd.read_csv(r'/data/workspace_files/CarPrice_Assignment.csv')
dataset.head()
```

```python
dataset.describe()
```

```python
# Quito campos indeseados
dataset=dataset.drop(columns=['car_ID','CarName'],axis=1)

# resumen de los datos
dataset.head()
```

```python
df = pd.DataFrame(dataset)

print(df["cylindernumber"].unique())
```

```python
# Ajusto valores de etiquetas
fueltype_ = { 'gas':1, 'diesel':2 }
dataset['fueltype'] = dataset['fueltype'].map(fueltype_)

aspiration_ = { 'std':1, 'turbo':2 }
dataset['aspiration'] = dataset['aspiration'].map(aspiration_)

doornumber_ = { 'two':1, 'four':2 }
dataset['doornumber'] = dataset['doornumber'].map(doornumber_)

carbody_ = { 'convertible':1, 'hatchback':2, 'sedan':3, 'wagon':4, 'hardtop':5 }
dataset['carbody'] = dataset['carbody'].map(carbody_)

drivewheel_ = { 'rwd':1, 'fwd':2, '4wd':3 }
dataset['drivewheel'] = dataset['drivewheel'].map(drivewheel_)

enginelocation_ = { 'front':1, 'rear':2 }
dataset['enginelocation'] = dataset['enginelocation'].map(enginelocation_)

enginetype_ = { 'dohc':1, 'ohcv':2, 'ohc':3, 'l':4, 'rotor':5, 'ohcf':6, 'dohcv':7 }
dataset['enginetype'] = dataset['enginetype'].map(enginetype_)

cylindernumber_ = { 'four':1, 'six':2, 'five':3, 'three':4, 'twelve':5, 'two':6, 'eight':7 }
dataset['cylindernumber'] = dataset['cylindernumber'].map(cylindernumber_)

fuelsystem_ = { 'mpfi':1, '2bbl':2, 'mfi':3, '1bbl':4, 'spfi':5, '4bbl':6, 'idi':7, 'spdi':8 }
dataset['fuelsystem'] = dataset['fuelsystem'].map(fuelsystem_)

# Resumen de los datos
dataset.head()
```

```python
# Divicion del dataset entre X(variables de entrada) y Y(variables de salida)
from matplotlib import pyplot as plt

Datos=np.array(dataset)

X = Datos[:,0:22]
Y = Datos[:,23]

#Normalizo datos
Y_N = ( Y-min(Y) )/( max(Y) - min(Y) )
plt.plot(Y_N)
```

```python
# Separar entre conjuntod de entrenamiento y testing
import sklearn 
from sklearn.model_selection import train_test_split

X_train, X_test,Y_train, Y_test= train_test_split(X, Y_N, test_size=0.2, random_state = 4296581)
print( X_train.shape, X_test.shape )
```

```python
# Normalizacion de los datos
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Entrenamiento del modelo de regresion lineal
from matplotlib import pyplot as plt
from sklearn import linear_model

Regr = linear_model.LinearRegression()
Regr.fit(X_train, Y_train)

 # Hacemos la predicción
Y_pred = Regr.predict(X_test)
plt.plot(Y_test,'b')
plt.plot(Y_pred,'r')
plt.show()
```

```python
#Calculando desempeño (Exactitud)
from sklearn.metrics import r2_score

print("Accuracy ",r2_score(Y_test, Y_pred))
```

```python
from keras import backend as K

X_train1 = K.cast_to_floatx(X_train)
y_train1 = K.cast_to_floatx(Y_train)
```

```python
#Realizando la regresión con una red neuronal
from keras.models import Sequential
from keras.layers import Dense

# create model
model = Sequential()
model.add(Dense(200, input_dim=22, activation='relu'))
model.add(Dense(150, activation='relu'))
model.add(Dense(140, activation='relu'))
model.add(Dense(130, activation='relu'))
model.add(Dense(120, activation='relu'))
model.add(Dense(110, activation='relu'))
model.add(Dense(100, activation='relu'))
model.add(Dense(90, activation='relu'))
model.add(Dense(80, activation='relu'))
model.add(Dense(70, activation='relu'))
model.add(Dense(60, activation='relu'))
model.add(Dense(50, activation='relu'))
model.add(Dense(40, activation='relu'))
model.add(Dense(30, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
np.random.seed(7)

# Compile model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Fit the model
model.fit(X_train1, y_train1, epochs=1000)
```
